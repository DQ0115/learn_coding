```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
data = pd.read_csv('C:/Users/user/Desktop/.csv')

# CSV
# 
years = data['Year']
attendances = data['Total_number_of_viewers']

# 
plt.figure(figsize=(10, 6))
plt.plot(years, attendances, marker='o', linestyle='-')

plt.title('Total number of live visitors of CPBL over the years')
plt.xlabel('Year')
plt.ylabel('number_Ten thousand')

plt.grid(True)
plt.xticks(years[::1])
plt.tight_layout()

# 
plt.show()
```
