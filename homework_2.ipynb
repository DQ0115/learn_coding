```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
data = pd.read_csv('C:/Users/user/Desktop/中華職棒大聯盟歷年現場觀眾總人次.csv')

# 假设CSV文件包含两列：年份和人数
# 如果列名不同，请适当更改
years = data['Year']
attendances = data['Total_number_of_viewers']

# 创建图表
plt.figure(figsize=(10, 6))
plt.plot(years, attendances, marker='o', linestyle='-')

plt.title('Total number of live visitors of CPBL over the years')
plt.xlabel('Year')
plt.ylabel('number_Ten thousand')

plt.grid(True)
plt.xticks(years[::1])
plt.tight_layout()

# 显示图表
plt.show()
```
