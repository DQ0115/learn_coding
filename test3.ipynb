```python
import pandas as pd
import numpy as np

# 讀取CSV檔案並顯示數據
data = pd.read_csv('C:/Users/user/Desktop/coding_class/adult.csv')
print(data)

# 使用NumPy計算薪資的平均值
average_Income = np.mean(data['Income'])
print("平均薪資:", average_Income)

# 使用NumPy計算薪資的標準差
std_Income = np.std(data['Income'])
print("薪資標準差:", std_Income)

# 使用NumPy計算薪資的中位數
median_Income = np.median(data['Income'])
print("薪資中位數:", median_Income)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Extract the salary and headcount columns from the DataFrame
Salary = data['Income']

# Create a histogram of the salaries using Matplotlib
plt.hist(Salary, bins=100)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Quantity')

plt.axvline(x=average_Income, color='red', linestyle='--')# 平均數指示線
plt.axvline(x=average_Income + std_Income, color='green', linestyle='--')# 平均數指示線 + 一個標準差
plt.axvline(x=average_Income - std_Income, color='green', linestyle='--')# 平均數指示線 - 一個標準差

plt.show()
```
